```python
bm25_terrier_stemmed = pt.BatchRetrieve.from_dataset('msmarco_passage', 'terrier_stemmed', wmodel='BM25')
```

```python
# Lets use a Vanilla BERT ranker from OpenNIR. We'll use the Capreolus model available from Huggingface
vanilla_bert = onir_pt.reranker('hgf4_joint', ranker_config={'model': 'Capreolus/bert-base-msmarco', 'norm': 'softmax-2'})
```

```python
bm25_terrier_stemmed_text = pt.BatchRetrieve.from_dataset('msmarco_passage', 'terrier_stemmed_text', wmodel='BM25', metadata=['docno', 'text'])

bm25_bert_terrier_stemmed_text = bm25_terrier_stemmed_text >> vanilla_bert
```

```python
bm25_terrier_stemmed_docT5query = pt.BatchRetrieve.from_dataset('msmarco_passage', 'terrier_stemmed_docT5query', wmodel='BM25')
```

```python
bm25_terrier_stemmed_deepct = pt.BatchRetrieve.from_dataset('msmarco_passage', 'terrier_stemmed_deepct', wmodel='BM25')
```

```python
pt.Experiment(
    [bm25_terrier_stemmed, bm25_terrier_stemmed_text, bm25_bert_terrier_stemmed_text, bm25_terrier_stemmed_docT5query, bm25_terrier_stemmed_deepct],
    dataset.get_topics('test-2019'),
    dataset.get_qrels('test-2019'),
    batch_size=200,
    filter_by_qrels=True,
    eval_metrics=['recip_rank','num_rel','num_ret','num_rel_ret','recall_10','recall_100','recall_1000','ndcg_cut_10','ndcg_cut_100'],
    names=['bm25_terrier_stemmed', 'bm25_terrier_stemmed_text', 'bm25_bert_terrier_stemmed_text', 'bm25_terrier_stemmed_docT5query', 'bm25_terrier_stemmed_deepct'])
```

##Comparing baseline BM25 worst performing top 5 queries on BM25+BERT

```python
from pyterrier.measures import *
dataset = pt.get_dataset("msmarco_passage")
topics = dataset.get_topics("test-2019")
qrels= dataset.get_qrels("test-2019")
#dataset.get_topics("test-2019").iloc[3]

exp_bm25_bert = pt.Experiment(
    [bm25_bert_terrier_stemmed_text],
    topics,
    qrels,
    filter_by_qrels=True,
    perquery=True,
    eval_metrics=['recip_rank','num_rel','num_ret','num_rel_ret','recall_10','recall_100','recall_1000','ndcg_cut_10','ndcg_cut_100'],
)
exp_bm25_bert
```

```python
import pandas as pd


exp = exp_bm25_bert.copy()
exp = exp.pivot_table('value', ['name', 'qid'], 'measure')

# exp.columns = exp.columns.droplevel(0) #remove amount
# exp.columns.name = None               #remove categories
exp = exp.reset_index()  
exp = exp.drop('name', 1)
exp = exp.sort_values(by=['ndcg_cut_10'])
# recall 1000 == metric below
# exp['rel_percent'] = exp['num_rel_ret']/exp['num_rel']
exp


```

```python
exp_bm25_vanilla = pt.Experiment(
    [bm25_terrier_stemmed],
    topics,
    qrels,
    filter_by_qrels=True,
    perquery=True,
    eval_metrics=['ndcg_cut_10'],
)
```

```python
bm25_bert_exp = pt.Experiment(
    [bm25_bert_terrier_stemmed_text],
    topics,
    qrels,
    filter_by_qrels=True,
    perquery=True,
    eval_metrics=['ndcg_cut_10'],
)
```

```python
exp_bm25_vanilla.head()
```

```python
compare_df = pd.DataFrame()
import numpy as np
compare_df['ndcg_improve'] = np.where(bm25_bert_exp['value'] > exp_bm25_vanilla['value'], 'True', 'False')
compare_df[compare_df.ndcg_improve == 'True'].shape[0]
#improved queries out of 43
```
